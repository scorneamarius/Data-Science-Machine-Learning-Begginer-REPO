```python
import pandas as pd
import os
import numpy as np
from matplotlib import pyplot as plt
```

In data directory, there are sales csv files for every month.
We need to concatenate them into one dataframe

```python
path = 'data'
for _, _, filenames in os.walk(path):
    continue
filenames
```

```python
data = pd.DataFrame()
for file in filenames:
    current_path = "data/" + file
    current_df = pd.read_csv(current_path)
    data = data.append(other = current_df)
```

```python
data.shape
```

```python
data.head(10)
```

Check for missing values

```python
columns = data.columns
for column in columns:
    print(f'{column} has {data[column].isna().sum()} missing values')
```

Drop the rows that contains at least one missing value

```python
data.dropna(axis = 0,how = 'any', inplace = True )
```

Check again for missing values

```python
columns = data.columns
for column in columns:
    print(f'{column} has {data[column].isna().sum()} missing values')
```

Let add a column only for month

```python
month_data = []
for date in data['Order Date'].to_numpy():
    month_data.append(date.split('/')[0])
month_data = np.array(month_data)
```

```python
unique, counts = np.unique(month_data, return_counts = True)
print(dict(zip(unique, counts)))
```

Also, we have observed that we have some problems with that column

```python
data.loc[data['Order Date'] == 'Order Date']
```

Delete these column

```python
data.drop(data[data['Order ID'] == 'Order ID'].index, inplace = True)
```

```python
data.loc[data['Order Date'] == 'Order Date']
```

```python
month_data = []
for date in data['Order Date'].to_numpy():
    month_data.append(date.split('/')[0])
month_data = np.array(month_data)
unique, counts = np.unique(month_data, return_counts = True)
print(dict(zip(unique, counts)))
```

Append month data column

```python
data['month'] = month_data
```

```python
data.head(10)
```

Let add a column where we can find the total prices sales for each entry

```python
data.info()
```

```python
data['Quantity Ordered'] = data['Quantity Ordered'].astype(dtype = np.int32)
data['Price Each'] = data['Price Each'].astype(dtype = np.float32)
```

```python
data.info()
```

```python
quantity = data['Quantity Ordered']
price = data['Price Each']
total_price = [i*j for (i,j) in zip(quantity,price)]
```

```python
data['total_price'] = total_price
data.head()
```

#### The total sales for each month

```python
price_per_month = data.groupby(by = 'month').sum()
price_per_month
```

```python
import seaborn
```

```python
seaborn.barplot(data = price_per_month, y = 'total_price',x = price_per_month.index)
```

#### What city does sold the most product?

```python
addresses = data['Purchase Address'].to_numpy()
cities = []
for address in addresses:
    interested = address.split(',')
    cities.append(interested[1] + " " + interested[2][:3])

cities_data = np.array(cities)
unique, counts = np.unique(cities_data, return_counts = True)
print(dict(zip(unique, counts)))
    
```

```python
data['city'] = cities_data
```

```python
sales_per_city = data.groupby(by = 'city').sum()
sales_per_city
```

```python
seaborn.barplot(data = sales_per_city, y = 'total_price', x = sales_per_city.index)
plt.xticks(rotation = 90)
plt.show()
```

#### What time should we display advertisements to maximize likelihood of customer's buying product?

```python
data.head()
```

```python
order_date = data['Order Date'].to_numpy()
```

```python
hours = []
for date in order_date:
    hours.append(date.split(' ')[1].split(':')[0])
```

```python
hours = np.array(hours)
unique, counts = np.unique(hours, return_counts = True)
dict(zip(unique,counts))
```

```python
data['hours'] = hours
data.head()
```

```python
prices_by_hour = data.groupby(by = 'hours').sum()
prices_by_hour
```

```python
index = prices_by_hour.index.to_numpy(dtype = np.int8)
index
```

```python
plt.plot(index,prices_by_hour['total_price'])
plt.xt
```
