The next 2 functions are used for compute dLoss/dw1 and dLoss/dw2  

```python
def w1_derivative(y, w1, w2):
    s = 0
    n = len(y)
    for i in range(n):
        s = -2 *i*(y[i] - (w1 * i + w2))
    return s / n
```

```python
def w2_derivative (y, w1 ,w2):
    s = 0
    n = len(y)
    for i in range(n):
        s = s -2 * (y[i] - (w1 * i + w2))
    return s / n
```

# Random guess for w1 and w2
    Suppose that our model is a linear regression y = w1*x + w2
    Our goal is to minimize the Loss function
    We need to find w1 and w2 for that Loss function has a minimum value
    In each epoch, we compute dLoss/dw1 and dLoss/dw2 and update the actual w1 and w2
        w1 = w1 - lr * dloss_dw1
        w2 = w2 - lr * dloss_dw2
    lr is a small number to ensure we don't jump over a minimum value 
    At the beginning, we assign w1 and w2 with a random values

```python
w1 = 2
w2 = 1
```

For actual w1 and w2, we calculate the y predicted

```python
preds = np.array([w1*i + w2 for i in x])
```

```python
plt.scatter(x,y)
plt.plot(x,preds)
plt.show()
```

The loss for our initial guess is following:

```python
mean_squared_error(y,w1,w2)
```

# Train loop
    Also, we keep the loss value for each epoch 

```python
epochs = 100
MSE = []
lr = 0.001
for _ in range(epochs):
    MSE.append(mean_squared_error(y,w1,w2))
    dloss_dw1 = w1_derivative(y, w1, w2)
    dloss_dw2 = w2_derivative(y, w1 ,w2)
    w1 = w1 - lr * dloss_dw1
    w2 = w2 - lr * dloss_dw2
    
```

# Compute the new y predicted with w1 and w2 adjusted

```python
preds2 = np.array([w1*i + w2 for i in x])
```

# Plot our outcomes

```python
fig,axs = plt.subplots(2,2,figsize=(15,10))
axs[0, 0].scatter(x,y)
axs[0,0].title.set_text('First Plot')
axs[0,0].set_xlabel('X')
axs[0,0].set_ylabel('Y')

axs[0,1].scatter(x,y)
axs[0,1].plot(x,preds)
axs[0,1].title.set_text('Random guess')
axs[0,1].set_xlabel('X')
axs[0,1].set_ylabel('Y')

axs[1,0].scatter(x,y)
axs[1,0].plot(x,preds2)
axs[1,0].title.set_text('Guess after optimization')
axs[1,0].set_xlabel('X')
axs[1,0].set_ylabel('Y')

axs[1,1].plot(list(range(epochs)), MSE)
axs[1,1].title.set_text('Loss function')
axs[1,1].set_xlabel('Epoch')
axs[1,1].set_ylabel('Loss')
```
