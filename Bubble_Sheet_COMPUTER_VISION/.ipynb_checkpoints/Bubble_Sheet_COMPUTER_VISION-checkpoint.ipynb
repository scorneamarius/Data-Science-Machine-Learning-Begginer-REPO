#### Import out libraries

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# read the image
image = cv2.imread("image.png")
```

### Plot the image
##### To plot our image, we are going to use matplotlib library. Because OpenCV works with images in BGR format and matplotlib works with images in RGB, all time when we want to plot our image, we have to convert her in RGB format. We will define our function for that

```python
def plot_image(image, format_image):
    img = image.copy()
    if format_image == "RGB":
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img)
        plt.show()
    elif format_image == 'GRAY':
        img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        plt.imshow(img, cmap = 'gray')
        plt.show()
```

```python
plot_image(image,"RGB")
```

#### Declare some variables

```python
questions = 5
answers = 5
correct_ans = [0,1,0,0,0] # correct_ans[i] = j ->  i = row, j = correct_answer ( column )
```

#### Start preprocessing the image to find all shapes by contour

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plot_image(image, 'GRAY')
```

#### Blur image to reduce noise and preserve the edges

```python
blur_img = cv2.GaussianBlur(gray_image, (5,5), 0)
plt.imshow(blur_img, cmap = 'gray')
```

#### Find out the edges using Canny edge detector

```python
edge_img = cv2.Canny(blur_img, 10, 70)
plt.imshow(edge_img, cmap = 'gray')
```

```python
contours, _ = cv2.findContours(edge_img, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
cv2.drawContours(image, contours, -1,(0,0,255),3)
plot_image(image, 'RGB')
```

#### Next step is to find out only the paper document

##### We will iterare on all contours and approximate if the shape is rectangle 

```python
rect_list = []
for c in contours :
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.02 * peri, True)
    if len (approx) == 4:
        rect_list.append(approx)
```

#### Draw contours with blue color for rectangle shapes

```python
cv2.drawContours(image, rect_list, -1,(255,0,0),3)
plot_image(image, 'RGB')
```

```python
print(f'We find {len(rect_list)} rectangle shapes')
```

#### We know that the area of document is largest than others. Also, contour of question is the 2nd and grade is 3rd

```python
area_for_each_rect = []
for i,c in enumerate(rect_list):
    area = cv2.contourArea(c)
    area_for_each_rect.append(area)
area_for_each_rect
```

```python
index_sorted = np.argsort(np.array(area_for_each_rect))
index_sorted
```

#### Sort area and get the contour for document, questions and grade 

```python
document_contour = rect_list[index_sorted[-1]]
questions_contour = rect_list[index_sorted[-3]]
grade_contour = rect_list[index_sorted[-5]]
```

```python
k = 0
n = 0
for j,row in enumerate(rows):
    bubbles = [row[:,i*c:(i+1)*c] for i in range(0,5)]
    correct_answer = correct_ans[k]
    scores = []
    for i,bubble in enumerate(bubbles):
        flat = bubble.flatten()
        scores.append(len(flat[flat > 0]))
    predicted_answer = scores.index(max(scores))
    if predicted_answer == correct_answer:
        n = n + 1
    k = k + 1
print(f'Number of correct respons: {n}')
```
